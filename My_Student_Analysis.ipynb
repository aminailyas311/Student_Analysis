```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("C:/Users/Aj/Music/exams.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.figure(figsize=(5, 5))

ax = sns.countplot(data=df, x="gender", palette={"male": "blue", "female": "pink"}, hue="gender")
for container in ax.containers:
    ax.bar_label(container, label_type='edge')
plt.title("Distribution of Gender")
plt.show()
```

```python
#So we analyzed that number of females in the data is more than the male
```

```python
df.columns
```

```python
gb=df.groupby("parent_education_level").agg({"math":"mean"})
print(gb)
```

```python
sns.heatmap(gb ,annot=True)
plt.title("Relationship between parent's Education and Student's Scores ")
plt.show()
```

```python
#from the above chart we have concluded that the education of parents have a good impact
```

```python
plt.figure(figsize=(6, 6))

ax = sns.countplot(data=df, x="test_prep_course", hue="gender")
for container in ax.containers:
    ax.bar_label(container, label_type='edge')
plt.title("Distribution of test_prep_course ")
plt.show()
```

```python

#from the above graph we can see that most of the female does not complete test preperation course
```

```python

plt.figure(figsize=(7, 5))
ax = sns.countplot(data=df, x="gender", hue="math", palette="coolwarm")
plt.title("Math scores Distribution")
plt.show()
```

```python

#from the above graph we see female has greater marks in math as compared to male
```

```python
plt.figure(figsize=(6, 6))

ax = sns.countplot(data=df, x="race/ethnicity",palette={"group A": "blue", "group B": "pink","group C" : "orange","group D":"green","group E":"red"}, hue="race/ethnicity")
for container in ax.containers:
    ax.bar_label(container, label_type='edge')
plt.title("Distribution of ethnicity")
plt.show()
```

```python
#in the above graph we see that group B has the highest numbers than others
```
